# 3. Data Cleaning

## Missing Values Treatment

```python
# columns with null values and their count
def finding_null_values(data):
    total_columns = data.columns.to_list()
    cols_with_nan = []
    for i in total_columns:
        if data[i].isna().sum()>0:
            cols_with_nan.append((i,data[i].isna().sum()))
    return cols_with_nan
```

```python
df_train_cols_with_nan = finding_null_values(df_train)
```

```python
print(df_train_cols_with_nan)
```

```python
# cleaning the null values
def cleanin_data(data):
    # replacing missing values with none because they don't have that's why we replacing it with none
    for col in ('PoolQC','MiscFeature','GarageType','Alley','Fence','FireplaceQu','GarageFinish',
               'GarageQual','GarageCond','MasVnrType'):
        data[col] = data[col].fillna('None')
        
    #Replacing missing value with 0(since no garage = no cars in such garage)
    data['GarageYrBlt'].fillna(0,inplace=True)
        
    # Replacing null with zero assuming it doesnot have area assuming it has 0
    data["MasVnrArea"] = data["MasVnrArea"].fillna(0)
    
    # for below categorical basement-related feature NaN means that there is no basement 
    for col in ('BsmtQual', 'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2'):
        data[col] = data[col].fillna('None')

     #group by Neigborhood and fill missing value with median Lot frontage of all the neighboorhood
    data["LotFrontage"] = data.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median())) 
    
    return data
```

```python
df_train = cleanin_data(df_train)
```

```python
# droping the row that have null in electrical column
df_train.dropna(subset=['Electrical'],inplace=True,axis=0)
```

```python
#droping col MiscFeature because it has 1406 missing values same with utilities
df_train.drop(['MiscFeature','Utilities'],axis=1,inplace=True)
```

```python
# creating function for label encoding
def label_encoding(data):
    from sklearn.preprocessing import LabelEncoder
    cols = ('MSZoning','FireplaceQu','GarageType','BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
            'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 
            'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
            'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 
             'LotConfig','LandContour','SaleType','SaleCondition',
           'Neighborhood','Condition1','Condition2','BldgType','HouseStyle','RoofStyle','RoofMatl','Exterior1st',
           'Exterior2nd','MasVnrType','Foundation','Heating','Electrical')

    for i in cols:
        lb = LabelEncoder()
        data[i]=lb.fit_transform(data[i])
        
    return data
```

# 4. Feature Engineering

# Feature Scaling

Feature scaling is a preprocessing technique used in machine learning to standardize or normalize the range of independent variables or features in the dataset. The goal of feature scaling is to ensure that all features have a similar scale, preventing features with larger magnitudes from dominating those with smaller magnitudes during model training.

## Standard Scaler

StandardScaler standardizes features by removing the mean and scaling to unit variance. It transforms the data such that each feature has a mean of 0 and a standard deviation of 1.

**Standardized Value = (Original Value - Mean of Feature)/(Standard Deviation of feature)**




##### x is the original value of the feature.

##### Î¼ is the mean of the feature.

##### Ïƒ is the standard deviation of the feature.

##### z is the standardized value.

# 5. Model Training

## Linear Regression

Linear regression is a statistical method used to model the relationship between a dependent variable (target) and one or more independent variables (predictors) by fitting a linear equation to observed data. The equation takes the form:

#### y=mx+b

Where:


y is the dependent variable (target).

x is the independent variable (predictor).

m is the slope of the line (coefficient).

b is the y-intercept.

## Assumption of linear regression

#### 1. Linearity

This assumption states that the relationship between the independent variables and the dependent variable should be linear. It implies that a change in the dependent variable should correspond to a proportional change in the independent variable(s).
Linear regression assumes that the true relationship between the variables can be adequately represented by a straight line. However, if the true relationship is nonlinear, linear regression may not provide accurate predictions.

#### 2. Normality of Residual

This assumption states that the residuals of the model should follow a normal distribution. In other words, the errors (residuals) should be normally distributed with a mean of zero.
Deviations from normality can indicate that the model is not capturing all the underlying patterns in the data, leading to biased estimates and inaccurate hypothesis testing.

### 3. Homoscadicity

Homoscedasticity refers to the assumption that the variance of the residuals (the differences between the observed and predicted values) should be constant across all levels of the independent variables.
In practical terms, homoscedasticity implies that the spread of the residuals should not systematically change as the values of the independent variables change. If the spread of the residuals increases or decreases with the independent variable, it indicates heteroscedasticity.

#### 4. Independence

Independence assumption means that the observations in the dataset should be independent of each other. In simpler terms, the value of one observation should not be influenced by the value of another observation.
Violation of independence can occur in various situations, such as time series data where observations are correlated over time or in experimental data where measurements taken from the same subject are correlated.

# 6. Fine Tunning

### What is Ordinary Least Squares
**Ordinary Least Squares (OLS)** is a method used in statistical modeling to estimate the parameters of a linear regression model. In simple terms, it's a technique to find the best-fitting line through a set of points in a scatterplot.

```python
# making a list of all columns 
x_columns = df_train.columns.to_list()[:-1]
y_columns = df_train.columns.to_list()[-1]
```

```python
x = df_train[x_columns]
y = df_train[y_columns]
result = sm.OLS(y,x).fit()
print(result.summary())
```

```python
def get_another_stats():
    x = df_train[x_columns]
    y = df_train[y_columns]
    results = sm.OLS(y, x).fit()
    
    # Combine parameters and p-values into a dictionary
    results_dict = dict(zip(x_columns, results.pvalues))
    
    return results_dict
```

```python
result_list = get_another_stats()
```

```python
flag = True
while flag:
    result_list = get_another_stats()
    col_name = ''
    maxi = 0
    for i, j in result_list.items():
        if j>maxi and j>0.05:
            col_name = i
            maxi = j
    if col_name in x_columns:
        x_columns.remove(col_name)
    if col_name == '':
        flag = False
```

```python
x_columns
```

```python
# checking p-values>0.05
x = df_train[x_columns]
y = df_train[y_columns]
results = sm.OLS(y, x).fit()
print(results.summary())
```
